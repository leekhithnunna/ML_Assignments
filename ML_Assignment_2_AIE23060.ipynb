Q1. Write a program to count the number of vowels and consonants present in an input string.

```python
def count(input_str):
    vowels = "aeiouAEIOU"
    vowel_count = input_str.count('a') + input_str.count('e') + input_str.count('i') + input_str.count('o') + input_str.count('u') + input_str.count('A') + input_str.count('E') + input_str.count('I') + input_str.count('O') + input_str.count('U')

    consonant_count = len(input_str) - vowel_count

    return vowel_count, consonant_count

def main():
    input_str = str(input("Enter the string: "))

    vowels, consonants = count(input_str)

    if vowels > 0:
        print("Number of vowels:", vowels)
    if consonants > 0:
        print("Number of consonants:", consonants)
main()
```

Q2. Write a program that accepts two matrices A and B as input and returns their product AB. Check if A & B are multipliable; if not, return error message

```python
def multiply(mat1, mat2):
    if len(mat1[0]) != len(mat2):
        raise ValueError("Matrix Dimensions are not matched")

    res = [[0 for x in range(len(mat2[0]))]
           for y in range(len(mat1))]

    for i in range(len(mat1)):
        for j in range(len(mat2[0])):
            for k in range(len(mat2)):
                res[i][j] += mat1[i][k] * mat2[k][j]
    return res

def main():
    rows1 = int(input("Enter the no.of rows for matrix 1: "))
    cols1 = int(input("Enter the no.of columns for matrix 1: "))
    mat1 = []
    print("Enter elements in matrix 1:")
    for i in range(rows1):
        row = list(map(int, input().split()))
        mat1.append(row)

    rows2 = int(input("Enter the no.of rows in matrix 1: "))
    cols2 = int(input("Enter the no.of columns in matrix 2: "))
    mat2 = []
    print("Enter elements in matrix 2:")
    for i in range(rows2):
        row = list(map(int, input().split()))
        mat2.append(row)

    try:
        result = multiply(mat1, mat2)
        print("Result:")
        for row in result:
            print(row)
    except ValueError as e:
        print("Error:", e)

main()
```

Q3. Write a program to find the number of common elements between two lists. The lists
contain integers.

```python
def findcommon(list1, list2):
    commonele = list(set(list1) & set(list2))

    return len(commonele), commonele

def main():
    list1_size = int(input("Enter the no.of elements in 1st list: "))
    list1 = []
    print("Enter elements in 1st list:")
    for _ in range(list1_size):
        ele = int(input())
        list1.append(ele)

    list2_size = int(input("Enter the no.of elements in 2nd list: "))
    list2 = []
    print("Enter elements in 2nd list:")
    for _ in range(list2_size):
        ele = int(input())
        list2.append(ele)

    commoncount, common_num = findcommon(list1, list2)

    print("No.of common elements:", commoncount)
    if commoncount > 0:
        print("Common elements:", common_num)

main()
```

Q4. Write a program that accepts a matrix as input and returns its transpose.

```python
def matrix_transpose(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    transpose = [[0 for _ in range(rows)] for _ in range(cols)]

    for i in range(rows):
        for j in range(cols):
            transpose[j][i] = matrix[i][j]

    return transpose

def main():
    rows = int(input("Enter the no.of rows: "))
    cols = int(input("Enter the no.of columns: "))

    print("Enter the matrix:")
    matrix = []
    for _ in range(rows):
        row = list(map(int, input().split()))
        matrix.append(row)

    result = matrix_transpose(matrix)

    print("Transpose matrix= ")
    for row in result:
        print(" ".join(map(str, row)))

main()
```
