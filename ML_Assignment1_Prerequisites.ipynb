Q1. Create a vector, V1, (length = 100) of random number. Sort them in increasing order.

```python
import numpy as ny

V1 = ny.random.rand(100)

sorted_v1 = ny.sort(V1)

print("Sorted Vector(V1):\n", sorted_v1)
```

Q2. Find out what happens when perform V1*3. Multiply V1 with 3 to scale each value 3 times.

```python
scaled_V1 = V1 * 3

print("Original Vector(V1):\n", V1)
print("\n")
print("Scaled Vector (V1 X 3):\n", scaled_V1)
```

Q3. Find the mean and standard deviation for the vector.

```python
mean_V1 = ny.mean(V1)
std_V1 = ny.std(V1)

print("Mean of Vector(V1):", mean_V1)
print("Standard Deviation of V1:", std_V1)
```

Q7. Taking the above string S3, split the string into words by using “white space” and “period(.)” as
splitters. Put the words into an array. Find the length of the array.

```python
S3 = "I am a great learner. I am going to have an awesome life. I work hard and shall be rewarded well"

words = list(filter(None, S3.replace('.', ' ').split()))

array_length = len(words)

print("Array of Words:\n", words)
print("Length of the Array:", array_length)
```

Q8. Remove words “I”, “Am”, “to” & “and” from the array. Also, remove words containing more than
6 characters. Find the length of the array.

```python
words = ["I", "am", "a", "great", "learner", "I", "am", "going", "to", "have", "an", "awesome", "life",
         "I", "work", "hard", "and", "shall", "be", "rewarded", "well"]

filtered = filter(lambda word: word not in ["I", "Am", "to", "and"] and len(word) <= 6, words)

length = len(list(filtered))

print("\nLength of Filtered Array:", length)
```

Q9. Consider the date “01-JUN-2021”. Split the date and find the values for date, month and year
separately. Find the numerical value for the month using calendar sequence (Jan = 1, Dec = 12).

```python
date = "01-JUN-2021"
day, month, year = date.split("-")

month_dict = {"JAN": 1, "FEB": 2, "MAR": 3, "APR": 4, "MAY": 5, "JUN": 6,
              "JUL": 7, "AUG": 8, "SEP": 9, "OCT": 10, "NOV": 11, "DEC": 12}

month_numeric = month_dict[month.upper()]
print("Day:", day, "Month:", month_numeric, "Year:", year)
```

Q10. Create an excel file with 3 columns. The 3 columns contain the names of city, state and PIN code.

Q12. Create another vector V2 containing the square of the number present in V1. Make a plot of both
V1 and V2.

```python
V2 = sorted_v1 ** 2
plt.plot(sorted_v1, label="V1", color='blue')
plt.plot(V2, label="V2", color='green')
plt.title("Comparison of V1 and V2")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.show()
```
