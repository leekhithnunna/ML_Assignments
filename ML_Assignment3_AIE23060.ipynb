Q1. Evaluate the intraclass spread and interclass distances between the classes in your dataset. If
your data deals with multiple classes, you can take any two classes. Steps below (refer below
diagram for understanding):

```python
import pandas as pd
import numpy as np

# Load the preprocessed dataset
file_path = "/content/dataset.csv"
df = pd.read_csv(file_path)

df.head()
```

Calculate the mean for each class (also called as class centroid)
(Suggestion: You may use numpy.mean() function for finding the average vector for all
vectors in a given class. Please define the axis property appropriately to use this function. EX:
feat_vecs.mean(axis=0))


```python
class_0 = df[df["label"] == 0].iloc[:, :-2]  # Excluding sentiment and label
class_1 = df[df["label"] == 1].iloc[:, :-2]

centroid_0 = np.mean(class_0, axis=0)
centroid_1 = np.mean(class_1, axis=0)

print("Centroid of Class 0 (Fake Reviews):", centroid_0[:10])
print("Centroid of Class 1 (Truthful Reviews):", centroid_1[:10])
```

Calculate spread (standard deviation) for each class
(Suggestion: You may use numpy.std() function for finding the standard deviation vector
for all vectors in a given class. Please define the axis property appropriately to use this
function.)

```python
spread_0 = np.std(class_0, axis=0)
spread_1 = np.std(class_1, axis=0)

print("Spread of Class 0 (Fake Reviews):", spread_0[:10])
print("Spread of Class 1 (Truthful Reviews):", spread_1[:10])
```

Calculate the distance between mean vectors between classes
(Suggestion: numpy.linalg.norm(centroid1 – centroid2) gives the Euclidean
distance between two centroids.)

```python
interclass_distance = np.linalg.norm(centroid_0 - centroid_1)

print("Euclidean Distance Between Class Centroids:", interclass_distance)
```

Q3. Take any two feature vectors from your dataset. Calculate the Minkwoski distance with r from 1
to 10. Make a plot of the distance and observe the nature of this graph.


```python
feature1 = "tfidf_100"
feature2 = "tfidf_200"

vec1 = df[feature1].values
vec2 = df[feature2].values

r_values = range(1, 11)
distances = [np.linalg.norm(vec1 - vec2, ord=r) for r in r_values]

plt.figure(figsize=(8, 5))
plt.plot(r_values, distances, marker='o', linestyle='-', color='b')
plt.xlabel("Minkowski Order (r)")
plt.ylabel("Minkowski Distance")
plt.title(f"Minkowski Distance Between {feature1} and {feature2}")
plt.grid(True)
plt.show()
```

Q8. Make k = 1 to implement NN classifier and compare the results with kNN (k = 3). Vary k from 1 to
11 and make an accuracy plot.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

k_values = range(1, 12)
accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
    print(f"Accuracy for k={k}: {acc:.4f}")

plt.figure(figsize=(8, 5))
plt.plot(k_values, accuracies, marker='o', linestyle='-', color='b')
plt.xlabel("k Value")
plt.ylabel("Accuracy")
plt.title("Accuracy vs. k in kNN Classifier")
plt.grid(True)
plt.show()

print(f"Accuracy with k=1 (NN): {accuracies[0]:.4f}")
print(f"Accuracy with k=3 (kNN): {accuracies[2]:.4f}")
```

Q9. Please evaluate confusion matrix for your classification problem. From confusion matrix, the
other performance metrics such as precision, recall and F1-Score measures for both training and test
data. Based on your observations, infer the models learning outcome (underfit / regularfit / overfit).

```python
from sklearn.metrics import confusion_matrix, classification_report

y_train_pred = knn.predict(X_train)
y_test_pred = knn.predict(X_test)

train_cm = confusion_matrix(y_train, y_train_pred)
test_cm = confusion_matrix(y_test, y_test_pred)

print("Confusion Matrix - Training Data:\n", train_cm)
print("Confusion Matrix - Test Data:\n", test_cm)

train_report = classification_report(y_train, y_train_pred)
test_report = classification_report(y_test, y_test_pred)

print("\nClassification Report - Training Data:\n", train_report)
print("\nClassification Report - Test Data:\n", test_report)

train_acc = knn.score(X_train, y_train)
test_acc = knn.score(X_test, y_test)

print(f"Training Accuracy: {train_acc:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")

if train_acc > test_acc + 0.1:
    print("The model is overfitting (high training accuracy, low test accuracy).")
elif train_acc < test_acc:
    print("The model might be underfitting (poor training performance).")
else:
    print("The model is well-fitted (training and test accuracy are close).")
```
